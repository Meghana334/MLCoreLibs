## Double Plot

```python
x = [1,2,3,4,5,6,7,8,9]
y = [10,2,3,6,5,8,5,1,7] 
x2 = [1,2,3,4,5,6,7,8,9]
y2 = [2,4,7,9,5,3,8,10,2]
plt.plot(x,y,'r')
plt.plot(x2,y2,'g')
plt.plot(x,y,'ro')
plt.plot(x2,y2,'go')
plt.title("Graph")
plt.xlabel("count")
plt.ylabel("reach")
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [10,2,3,6,5,8,5,1,7] 
x2 = [1,2,3,4,5,6,7,8,9]
y2 = [2,4,7,9,5,3,8,10,2]

plt.plot(x,y,'r',label = "Profit")
plt.plot(x2,y2,'g',label = "Tax")

plt.plot(x,y,'ro')
plt.plot(x2,y2,'go')

plt.legend()
plt.title("Graph")
plt.xlabel("count")
plt.ylabel("reach")
plt.show()
```
